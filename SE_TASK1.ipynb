```python
import matplotlib.pyplot as plt
import numpy as np
def quadratic_model(time):
  #Hard coded coefficents for the quadratic equations
  a=0.2
  b=-5
  c=50

  #Quadratic equation represents temperature change
  temperature = a*(time**2)+b*time+c
  return temperature

def main():
  #Time values from 0 to 10(for example)
  time_values = np.linspace(0,10,50)

  #calculate temperatures using hard coded coefficients
  temperature_hardcoded = quadratic_model(time_values)

  #Plot the result
  plt.plot(time_values,temperature_hardcoded,label='Hard-coded Coefficients')
  plt.xlabel('Time')
  plt.ylabel('Temperature')
  plt.legend()
  plt.title('Weather modelling with Quadratic equation(hard-coded coefficient)')
  plt.show()
if_name_="main"
main()
```

```python
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def read_coefficients_from_file(filename):
    with open(filename, 'r') as f:
        coefficients = [float(line.strip()) for line in f]
    return coefficients

def main():
    time_values = np.linspace(0, 10, 50)

    # Read coefficients from file
    try:
        coefficients = read_coefficients_from_file('coefficients.txt')
        if len(coefficients) != 3:
            raise ValueError("The coefficients file should contain exactly three lines.")
        a, b, c = coefficients
    except FileNotFoundError:
        print("Error: coefficients.txt not found.")
        return
    except ValueError as e:
        print(f"Error: {e}")
        return

    temperature_from_file = quadratic_model(time_values, a, b, c)

    plt.plot(time_values, temperature_from_file, label='Coefficients from File')
    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation (Coefficients from File)')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    # Define multiple sets of parameters
    parameters_sets = [
        {'a': 1.0, 'b': 2.0, 'c': 3.0},
        {'a': 0.5, 'b': 1.0, 'c': 2.0},
        {'a': 2.0, 'b': -1.0, 'c': 5.0}
    ]

    # Plot each set of parameters
    for params in parameters_sets:
        a = params['a']
        b = params['b']
        c = params['c']

        # Calculate temperatures
        temperature_calculated = quadratic_model(time_values, a, b, c)

        # Plot the calculated temperatures
        plt.plot(time_values, temperature_calculated, label=f'a={a}, b={b}, c={c}')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation (Multiple Lines)')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    # Read parameters from file
    with open('parameters.txt', 'r') as file:
        for line in file:
            # Split the line and convert to float
            a, b, c = map(float, line.strip().split())

            # Calculate temperatures
            temperature_calculated = quadratic_model(time_values, a, b, c)

            # Plot each set of parameters
            plt.plot(time_values, temperature_calculated, label=f'a={a}, b={b}, c={c}')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation Multiple (File Input)')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    # Hardcoded coefficients
    a_hardcoded = 1
    b_hardcoded = -2
    c_hardcoded = 5

    # User input coefficients
    a_user = float(input("Enter the value of a: "))
    b_user = float(input("Enter the value of b: "))
    c_user = float(input("Enter the value of c: "))

    # Calculate temperatures using both sets of coefficients
    temperature_hardcoded = quadratic_model(time_values, a_hardcoded, b_hardcoded, c_hardcoded)
    temperature_user = quadratic_model(time_values, a_user, b_user, c_user)

    # Plot both models
    plt.plot(time_values, temperature_hardcoded, label='Hardcoded Coefficients: a=3,b=4,c=6')
    plt.plot(time_values, temperature_user, label='User Input Coefficients a=5,b=2,c=3', linestyle='--')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation')
    plt.show()

if __name__ == "__main__":
    main()
```
